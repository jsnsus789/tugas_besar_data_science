Gunakan Metode Newton

Define input problem  $Î²0=[0,0.01,0.01]$ ,  $TOL=10^{-5}$  and  $N=100$  iterations

```python
Beta = [0. ,0.01, 0.01]
TOL = 10**(-5)
N = 100
```

```python
# Nyatakan dalam bentuk Matriks

dfy = Y_test
dfx = X_test
dfx['ones'] = np.ones(len(dfy))
A = dfx.reindex(columns=["ones","TARGET","AMT_ANNUITY"])
A = A.to_numpy()

#print(A)
Y = np.transpose(dfy.to_numpy()).flatten()
print(Y)
```

```python
# Definisikan Galat

def error(a,b):
  return np.min(np.absolute(a-b))
```

```python
# Definisikan Fungsi Sigmoid

def Pi_func(A,B):
  Z= A@B
  return np.exp(Z)/(1+np.exp(Z))
```

```python
# Solution of Newton's Method

#step 1
k= 1
Beta0=np.ones(len(Beta)) #pilih nilai sembarang tidak sama dengan x untuk lolos pada loop while

print('iter', 'x1  x2', 'error')
print(k-1,Beta, error(Beta,Beta0))

#Step 2
while (error(Beta,Beta0)>TOL) and (k < N):
  Pr = Pi_func(A,Beta)
  var = Pr*(np.ones(len(Pr))-Pr)
  V = np.diag(var)
  #step 3
  Fx = -A.T@(Y-Pr)
  Jac = A.T@V@A
  #step 4
  term = np.linalg.inv(Jac)@Fx
  Beta0 = Beta
  #step 5
  Beta = Beta0 - term
  #step 6
  k = k + 1
  print(k - 1,Beta, error(Beta,Beta0))
```

```python
# Define Class Prediction

# A disini adalah A testing
Pr = Pi_func(A,Beta)
Y_predic = np.array([])
for a in Pr:
  if a <=0.5:
    Y_predic = np.append(Y_predic, 0)
  else:
     Y_predic = np.append(Y_predic, 1)

print(Y_predic)
```

```python
# Confusion Matrix

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y, Y_predic))
```

```python
print(classification_report(Y, Y_predic))
```
